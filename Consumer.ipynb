```python
from kafka import KafkaConsumer as kc

consumer = kc("mensagens",bootstrap_servers="127.0.0.1:9092",consumer_timeout_ms=1000, group_id="consumidores")
```

```python
for mensagem in consumer:
    print("Topic: ",mensagem.topic)
    print("Partition: ", mensagem.partition)
    print("Chave: ", mensagem.key)
    print("Offset: ",mensagem.offset)
    print("Mensagem: ",mensagem.value)
    print("---------------------")
```
